```python
from bs4 import BeautifulSoup
import pandas as pd

def scrapper(file_name):
    dic_files={"afr_ind":"data_py_pro/Africa_industries.aspx",
           "colorado":"data_py_pro/colorado.aspx",
           "jet_con":"data_py_pro/Jet_contractors.aspx",
           "lafarge":"data_py_pro/Lafarge_holcim.aspx",
           "tgcc":"data_py_pro/Tgcc.aspx"}
#
    with open(dic_files[file_name],"r") as file1:  # we used a dictionary to store files names and file paths, we assign
        globals()[file_name]=file1.read()           # a file with the "with open() as" so it closes automatically
        final_file=BeautifulSoup(globals()[file_name],'lxml')  # we store the content in final_file fulsoup to get the final page

    list_of_columns = ["date","closing","adjusted","evolution","quantity","volume"] #list of columns labels
    data=[]
    for i in range(0,733,6): # number of the days existing in data is 733, 6 jump lap is to go to date case again
        data.append([final_file.find_all("span")[i].get_text(),
        float(final_file.find_all("span")[i+1].get_text().replace(",",".")),
        float(final_file.find_all("span")[i+2].get_text().replace(",",".")),
        float(final_file.find_all("span")[i+3].get_text().replace(",",".")),
        float(final_file.find_all("span")[i+4].get_text().replace(",",".")),
        float(final_file.find_all("span")[i+5].get_text().replace(",","."))])

    data_df= pd.DataFrame(data,columns=list_of_columns)        
    data_csv= data_df.to_csv("{}.csv".format(file_name))
    return data_csv

```
